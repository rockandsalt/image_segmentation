```python
s1_path = "./data/S02_bined.csv"
s2_path = "./data/S18_bined.csv"
```

```python
%matplotlib inline
from rpy2.robjects import FloatVector
from rpy2.robjects.packages import importr
from numpy.fft import rfft
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


s1 = pd.read_csv(s1_path, sep=',',header=None).values
s2 = pd.read_csv(s2_path, sep=',',header=None).values
```

```python
base = importr('base')

s1_t = np.real(rfft(s1[0]))
s2_t = np.real(rfft(s2[0]))

fig,ax = plt.subplots()
ax.plot(range(s1_t.shape[0]-2),s1_t[2:])
ax.plot(range(s2_t.shape[0]-2),s2_t[2:])
plt.show()
```
